```python
from os import truncate
data_valida = False

while(not data_valida):
  data = input("Digite uma data: (dd/mm/aaaa)")
  dia, mes, ano = data.split("/")

  dia = int(dia)
  mes = int(mes)
  ano = int(ano)

  diaValido = True
  mesValido = True
  anoValido = True

  if(dia < 1 or dia > 31):
      print("Dia inválido")
      diaValido = False
  if(mes < 1 or mes > 12):
      print("Mês inválido")
      mesValido = False
  if(ano < 1 or ano > 2026):
      print("Ano inválido")
      anoValido = False

  if(diaValido and mesValido and anoValido):
    print(data, "é uma data válida")
    data_valida = True
  else:
    print(data, "é uma data inválida")
```

```python
while(True):
  nome_completo = input("Digite seu nome completo: ")
  palavras_nome = nome_completo.split(' ')

  if(len(palavras_nome) >= 2):
    print("Nome: "+ palavras_nome[0])
    tamanho = len(palavras_nome)
    email = (palavras_nome[0] + "." + palavras_nome[tamanho - 1] + "@ufn.edu.br").lower()
    print(email)
    break

  print("Nome inválido")
```

```python
frase = input("Digite uma frase: ")
print("A frase possui", len(frase), "caracteres")
print("Frase original: ", frase)
print("A frase em MAIÚSCULO: ", frase.upper())
print("A frase em minúsculo:", frase.lower())

##trocar os 'a' por @ e os 'e' por &

for i in range(len(frase)):
  if frase[i] == 'a':
    frase = frase.replace('a', '@')
  if frase[i] == 'e':
    frase = frase.replace('e', '&')

print("Frase modificada: ", frase)
```
