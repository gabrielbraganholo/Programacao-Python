```python
nome = input("Digite seu nome: ")
nota1b = float(input("Digite a nota do 1º bimestre: "))
nota2b = float(input("Digite a nota do 2º bimestre: "))
nota3b = float(input("Digite a nota do 3º bimestre: "))
nota4b = float(input("Digite a nota do 4º bimestre: "))

media = (nota1b+nota2b+nota3b+nota4b)/4

print(media)
if(media >= 7):
  print("Aprovado")
if(media >= 5 and media < 7):
  print("Recuperação")
if(media < 5):
  print("Reprovado")
```

```python
nome = input("Digite o nome do paciente: ")
idade = int(input("Digite a idade do paciente: "))
peso = float(input("Digite o peso do paciente: "))
altura = float(input("Digite a altura do paciente: "))

IMC = peso/(altura**2)

print(f"Paciente: {nome}\nIdade: {idade}\nIMC: {IMC:.2f}")
```

```python
nome = input("Digite o nome do funcionário: ")
base_h = float(input("Digite o valor base da hora de trabalho: "))
horas_t = float(input("Digite o número de horas trabalhadas por semana: "))
anos = int(input("Digite o ano de ingresso na impresa: "))

salario = base_h*(horas_t*4)

print(f"Salário do mês: R${salario:.2f}")

if(4 <= anos <= 8):
  salario = salario/0.04
if(8 < anos <= 12):
  salario = salario/0.06
if(anos > 12):
  salario = salario/0.08

print(f"Devido aos seus anos de empresa, seu novo salário será: R${salario:.2f}")
```
